Step 2: Handle missing values using mean/median/imputation

```python
# calculated the median value
median_age = int(df['Age'].median())
median_age
```

```python
# filled null values
df['Age'] = df['Age'].fillna(median_age)
```

```python
# converted from float to int 
df['Age'] = df['Age'].round().astype(int)
```

```python
df['Age'].isnull().sum()
```

```python
df['Age']
```

```python
df['Embarked']=df['Embarked'].bfill()
```

```python
# Adding column of Family
df['Family'] = df['SibSp']+df['Parch']
df.sample(15)
```

# 3. Visualization

```python
# Histogram
cols = ['Survived','Pclass','Age','Fare','Family']
df_num = df[cols]
df_num.hist(figsize=(10, 8))
plt.suptitle('Histograms of Numeric Features')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for i, col in enumerate(df_num, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]


for col in df_num:
    df = remove_outliers_iqr(df, col)


print("Shape after outlier removal:", df.shape)
```

```python
print("After Removing Outliers")
numerical_cols = ['Age', 'Fare']
plt.figure(figsize=(15, 8))
for i, col in enumerate(df_num, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

# Survival Rate for Children (Age < 16)

```python
# Create 'IsChild' column
df['IsChild'] = df['Age'] < 16

survival_by_age_group = df.groupby('IsChild')['Survived'].mean()
print("Survival rate (children vs. adults):\n", survival_by_age_group)


```

```python

plt.figure(figsize=(6, 4))
sns.barplot(x='IsChild', y='Survived', data=df)
plt.title('Survival Rate: Children vs. Adults')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Adult', 'Child'])
plt.ylim(0, 1)
plt.show()
```

```python
# Already created 'Family' = SibSp + Parch
# Create 'FamilySizeGroup'
df['FamilySizeGroup'] = df['Family'].apply(
    lambda x: 'Large' if x >= 3 else ('Small' if x > 0 else 'Alone')
)

# Survival rate by family size group
survival_by_family = df.groupby('FamilySizeGroup')['Survived'].mean()
print("Survival rate by family size:\n", survival_by_family)

```
